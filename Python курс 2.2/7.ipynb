```python
import pyttsx3
import speech_recognition as sr
import time
import os
import webbrowser
import random
import sys
engine = pyttsx3.init()  #engine = tts
voices = engine.getProperty('voices')
engine.setProperty('voice','ru')

for voice in voices:
    if voice.name == 'Victoria':
        engine.setProperty('voice', voice.id)

        
        
        
def record_volume():
    r = sr.Recognizer()
    print('Настраиваюсь.')
    engine.say('Жду')
    engine.runAndWait()
    with sr.Microphone(device_index = 2) as source:
        r.adjust_for_ambient_noise(source, duration=0.5) #настройка посторонних шумов
        print('Слушаю')
        audio = r.listen(source)
        print('Услышала.')
        try:
            s = r.recognize_google(audio, language = 'ru-RU').lower()
            print(s)
        except:
            print('Error')
        return s

    
def makeSomething(s):
    stringList = s.split()
    print(stringList)
    print(stringList[1:])
    tmp = ""
    for t in stringList[1:]:
        tmp += t + "+"
    print(tmp)
    if  'открой сайт GITHub' in s:
        engine.say('Открываю сайт GITHub.')
        engine.runAndWait()
        webbrowser.open('https://github.com/')   
    elif 'произнеси' in s or 'скажи' in s or 'повтори' in s:
        engine.say(stringList[10:])
        engine.runAndWait()
        print('Bot: ' + s[10].upper() + s[11:])
        
    elif 'случайное число' in s:
        ot=s.find('от')
        do=s.find('до')
        f_num=int(s[ot+3:do-1])
        l_num=int(s[do+3:])
        r=str(random.randint(f_num, l_num))
        print('Bot: ' + r)
        engine.say(r)
        engine.runAndWait()
    elif 'блокнот' in s:
        engine.say('Открываю блокнот')
        engine.runAndWait()
        os.system("notepad.exe")
    elif 'калькулятор' in s:
        engine.say('Открываю калькулятор')
        engine.runAndWait()
        os.system("calc.exe")
    elif stringList[0]=='найди':
        engine.say('Ищу')
        engine.runAndWait()
        os.system("start www.google.com/search?q=" + tmp)
    elif 'пока' in s or 'до свидания' in s:
        print('Bot: До свидания!')
        engine.say('До свидания')
        engine.runAndWait()
        sys.exit()
        
while True:
    makeSomething(record_volume())

    
```
